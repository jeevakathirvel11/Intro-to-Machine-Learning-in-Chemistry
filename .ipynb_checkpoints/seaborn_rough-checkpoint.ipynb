# Box Plot

```python
plt.figure(figsize=(14,9))
sns.set_style('darkgrid')

# A box plot allows you to compare different variables
# The box shows the quartiles of the data. The bar in the middle is the median and
# the box extends 1 standard deviation from the median
# The whiskers extend to all the other data aside from the points that are considered
# to be outliers
# Hue can add another category being sex
# We see men spend way more on Friday versus less than women on Saturday
sns.boxplot(x='day',y='total_bill',data=tips, hue='sex')

# Moves legend to the best position

plt.legend(loc=0)
```

# Violin Plot

```python
# Violin Plot is a combination of the boxplot and KDE
# While a box plot corresponds to data points, the violin plot uses the KDE estimation of the data points
# Split allows you to compare how the categories compare to each other
sns.violinplot(x='day',y='tip',data=tips, hue='sex',split=True, ) #dodge=True
```

# Strip Plot

```python
plt.figure(figsize=(8,5))

# The strip plot draws a scatter plot representing all data points where one
# variable is categorical. It is often used to show all observations with
# a box plot that represents the average distribution
# Jitter spreads data points out so that they aren't stacked on top of each other
# Hue breaks data into men and women
# Dodge separates the men and women data
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex', dodge=True)
```

# Scatter plot/Regression of random numbers - error bars

```python
x = np.random.normal(0, 1, 50)
y = x * 2 + np.random.normal(0, 2, size=x.size)

# Regression plot with error bars (confidence interval)
sns.regplot(x=x, y=y, ci=68)  # 'ci' sets the confidence interval (default is 95)
plt.show()
```

# Matplotlib

```python
import random
#Generate 20 random numbers between 10 and 100
randomlist_x = random.sample(range(1, 60), 20)

#Generate 20 random numbers between 1 and 100
randomlist_y = random.sample(range(1, 60), 20)

x = randomlist_x
y = randomlist_y


plt.scatter(x, y, label="scatter")

z = np.polyfit(x, y, 1) # z gives the slope and intercept as it's output on a list
p = np.poly1d(z) #p will give the "y" values correponding to the x values.
plt.plot(x,p(x),"r-", label="trendline")

plt.title("Scatter Plot and trendline")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.show()
```

```python


# Generate 20 random numbers between 10 and 100
randomlist_x = random.sample(range(1, 60), 20)
# Generate 20 random numbers between 1 and 100
randomlist_y = random.sample(range(1, 60), 20)

x = randomlist_x
y = randomlist_y

plt.scatter(x, y, label="scatter")
plt.xlabel("x values")
plt.ylabel("y values")

# Linear regression
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), "r-", label="trendline")

# Calculate residuals and split into upper/lower errors
residuals = p(x) - y
lower_errors = np.where(residuals < 0, -residuals, 0)  # Positive values for downward bars
upper_errors = np.where(residuals > 0, residuals, 0)    # Positive values for upward bars

# Plot residuals as asymmetric error bars
plt.errorbar(x, y, yerr=[lower_errors, upper_errors], fmt='none', ecolor='gray', alpha=0.7, label="residuals")

# Font settings
plt.rc('font', size=18)
plt.rc('axes', titlesize=18)
plt.rc('axes', labelsize=18)
plt.rc('xtick', labelsize=18)
plt.rc('ytick', labelsize=18)
plt.rc('legend', fontsize=14)
plt.rc('figure', titlesize=22)

plt.legend()
plt.show()
print(f"The curve fit is y = {z[0]:1.4E}x + {z[1]:1.4E}")
```

Equation for trend linear line

```python
from scipy import stats

# get coefficients of linear fit
slope, intercept, r_value, p_value, std_err = stats.linregress(df[df['dataset']=='I'].x,df[df['dataset']=='I'].y)

# use line_kws to set line label for legend
plot = sns.regplot(x="x", y="y", data=df[df['dataset']=='I'], ci=None, line_kws={'label':"y = {0:.1f} x + {1:.1f}; r = {2:.2f}".format(slope,intercept, r_value)}, scatter_kws={"s": 80})

# plot legend
plot.legend()
```

In order to visualize the linear regression parameters individually with scipy, we can calculate and print them. Here we will also calculate the R squared, p-value and standard error.


```python
# get coefficients of linear fit
slope, intercept, r_value, p_value, std_err, = stats.linregress(df[df['dataset']=='I'].x,df[df['dataset']=='I'].y)
print("slope: %.1f intercept: %.1f Rsquared:  %.2f  p-value: %f standard error: %f " % (slope, intercept,r_value, p_value, std_err))
print("slope:" + str(intercept))
```
